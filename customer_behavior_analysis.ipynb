```python
#Factors Influencing Customer Engagement
sql_1="""SELECT
    e.ContentType,
    p.Category,
    AVG(e.Likes) AS AvgLikes,
    AVG(CAST(SUBSTRING_INDEX(e.ViewsClicksCombined, '-', 1) AS UNSIGNED)) AS AvgViews,
     #Extracts the portion before the first `-` (Views), converts it to a number, and calculates the average
    AVG(CAST(SUBSTRING_INDEX(e.ViewsClicksCombined, '-', -1) AS UNSIGNED)) AS AvgClicks
     #Extracts the portion after the last `-` (Clicks), converts it to a number, and calculates the average
    FROM engagement_data AS e
    LEFT JOIN products AS p ON e.ProductID = p.ProductID
    GROUP BY e.ContentType, p.Category
    ORDER BY AvgLikes DESC, AvgViews DESC, AvgClicks DESC"""

# fetch all the selected records and converts them into dataframe
my_cursor.execute(sql_1)
Factors_customer_engagememt=my_cursor.fetchall()
columns = [col[0] for col in my_cursor.description]
Factors_customer_engagememt=pd.DataFrame(Factors_customer_engagememt,columns=columns)
Factors_customer_engagememt
```

```python
#Stages Where Customers Drop Off from Their Journey
sql_2="""SELECT
    stage,count(*) as stage_count,avg(Duration) as duration
    FROM customer_journey
    WHERE Action ='drop-off'
    GROUP BY stage
    ORDER BY duration DESC;"""

my_cursor.execute(sql_2)
Action_stages=my_cursor.fetchall()
columns = [col[0] for col in my_cursor.description]
Action_stages=pd.DataFrame(Action_stages,columns=columns)
Action_stages
```

```python
# Impact of Customer Reviews on Purchasing Behavior
sql_3="""SELECT
    p.ProductName,
    AVG(cr.Rating) AS AvgRating,
    COUNT(DISTINCT CASE WHEN cj.Action = 'Drop-off' THEN cj.CustomerID END) AS DropOffCount,
    COUNT(DISTINCT CASE WHEN cj.Action = 'Purchase' THEN cj.CustomerID END) AS PurchaseCount
    FROM customer_reviews AS cr
    LEFT JOIN customer_journey AS cj ON cr.ProductID = cj.ProductID
    LEFT JOIN products AS p ON cr.ProductID = p.ProductID
    GROUP BY p.ProductName
    ORDER BY AvgRating DESC, PurchaseCount DESC"""

my_cursor.execute(sql_3)
Impact_customer_review=my_cursor.fetchall()
columns = [col[0] for col in my_cursor.description]
Impact_customer_review=pd.DataFrame(Impact_customer_review,columns=columns)
Impact_customer_review
```

```python
# Top-Performing Products
sql_4="""SELECT
    p.ProductName,
    AVG(e.Likes) AS AvgLikes,
    COUNT(cj.CustomerID) AS PurchaseCount
    FROM products AS p
    LEFT JOIN engagement_data AS e ON p.ProductID = e.ProductID
    LEFT JOIN customer_journey AS cj ON p.ProductID = cj.ProductID
    WHERE cj.Action = 'Purchase'
    GROUP BY p.ProductName
    ORDER BY PurchaseCount DESC, AvgLikes DESC"""
my_cursor.execute(sql_4)
top_performing_products=my_cursor.fetchall()
columns = [col[0] for col in my_cursor.description]
top_performing_products=pd.DataFrame(top_performing_products,columns=columns)
top_performing_products
```

```python
# Top-Performing Locations
sql_5="""SELECT
    g.Country,
    g.City,
    COUNT(c.CustomerID) AS CustomerCount,
    AVG(e.Likes) AS AvgLikes
    FROM geography AS g
    LEFT JOIN customers AS c ON g.GeographyID = c.GeographyID
    LEFT JOIN engagement_data AS e ON c.CustomerID = e.ProductID
    GROUP BY g.Country, g.City
    ORDER BY CustomerCount DESC, AvgLikes DESC"""
my_cursor.execute(sql_5)
top_performing_locations=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
top_performing_locations=pd.DataFrame(top_performing_locations,columns=columns)
top_performing_locations
```

```python
#Top-Performing Customer Segments
sql_6="""SELECT
    c.Gender,
    CASE
        WHEN c.Age < 25 THEN 'Youth'
        WHEN c.Age BETWEEN 25 AND 40 THEN 'Adult'
        WHEN c.Age > 40 THEN 'Senior'
    END AS AgeSegment,
    COUNT(c.CustomerID) AS CustomerCount,
    AVG(e.Likes) AS AvgLikes
    FROM customers AS c
    LEFT JOIN engagement_data AS e ON c.CustomerID = e.ProductID
    GROUP BY Gender, AgeSegment
    ORDER BY CustomerCount DESC, AvgLikes DESC"""
my_cursor.execute(sql_6)
top_performing_customer_segement=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
top_performing_customer_segement=pd.DataFrame(top_performing_customer_segement,columns=columns)
top_performing_customer_segement
```

```python
#Analyze Customer Segments of the Reviews
sql_7="""SELECT
    c.Gender,
    CASE
        WHEN c.Age < 25 THEN 'Youth'
        WHEN c.Age BETWEEN 25 AND 40 THEN 'Adult'
        WHEN c.Age > 40 THEN 'Senior'
    END AS AgeSegment,
    g.Country,
    COUNT(cr.ReviewID) AS ReviewCount,
    AVG(cr.Rating) AS AvgRating
    FROM customer_reviews AS cr
    LEFT JOIN customers AS c ON cr.CustomerID = c.CustomerID
    LEFT JOIN geography AS g ON c.GeographyID = g.GeographyID
    GROUP BY Gender, AgeSegment, g.Country
    ORDER BY ReviewCount DESC, AvgRating DESC"""
my_cursor.execute(sql_7)
customer_segment_reviews=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
customer_segment_reviews=pd.DataFrame(customer_segment_reviews,columns=columns)
customer_segment_reviews
```

```python
#Pattern Between Negative Reviews and Product Performance
sql_8="""SELECT
    p.ProductName,
    AVG(cr.Rating) AS AvgRating,
    COUNT(DISTINCT CASE WHEN cr.Rating <= 2 THEN cr.CustomerID END) AS NegativeReviewCount,
    COUNT(DISTINCT CASE WHEN cj.Action = 'Purchase' THEN cj.CustomerID END) AS PurchaseCount,
    COUNT(DISTINCT CASE WHEN cj.Action = 'Drop-off' THEN cj.CustomerID END) AS DropOffCount
    FROM customer_reviews AS cr
    LEFT JOIN customer_journey AS cj ON cr.ProductID = cj.ProductID
    LEFT JOIN products AS p ON cr.ProductID = p.ProductID
    GROUP BY p.ProductName
    ORDER BY NegativeReviewCount DESC, DropOffCount DESC"""
my_cursor.execute(sql_8)
Negative_review_product_performance=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
Negative_review_product_performance=pd.DataFrame(Negative_review_product_performance,columns=columns)
Negative_review_product_performance
```

```python
##### Recommended startgies
# Improve Product Quality and Features
sql1_1="""SELECT
    p.ProductName,
    AVG(cr.Rating) AS AvgRating,
    COUNT(DISTINCT CASE WHEN cj.Action = 'Drop-off' THEN cj.CustomerID END) AS DropOffCount
    FROM customer_reviews AS cr
    LEFT JOIN customer_journey AS cj ON cr.ProductID = cj.ProductID
    LEFT JOIN products AS p ON cr.ProductID = p.ProductID
    GROUP BY p.ProductName
    HAVING AvgRating < 3
    ORDER BY DropOffCount DESC;"""
my_cursor.execute(sql1_1)
Improve_product_quality=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
Improve_product_quality=pd.DataFrame(Improve_product_quality,columns=columns)
Improve_product_quality
```

```python
#Customer Journey to Reduce Drop-Offs
sql1_2="""SELECT
    Stage,
    COUNT(*) AS DropOffCount
    FROM customer_journey AS cj
    WHERE Action = 'Drop-off'
    GROUP BY Stage
    ORDER BY DropOffCount DESC;"""
my_cursor.execute(sql1_2)
customer_journey_drop_off=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
customer_journey_drop_off=pd.DataFrame(customer_journey_drop_off,columns=columns)
customer_journey_drop_off
```

```python
#Enhance Customer Engagement Through Personalization
sql1_3="""SELECT
    c.Gender,
    CASE
        WHEN c.Age < 25 THEN 'Youth'
        WHEN c.Age BETWEEN 25 AND 40 THEN 'Adult'
        ELSE 'Senior'
    END AS AgeSegment,
    AVG(e.Likes) AS AvgLikes
    FROM customers AS c
    LEFT JOIN engagement_data AS e ON c.CustomerID = e.ProductID
    GROUP BY Gender, AgeSegment
    ORDER BY AvgLikes DESC"""
my_cursor.execute(sql1_3)
customer_performing_segements=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
customer_performing_segements=pd.DataFrame(customer_performing_segements,columns=columns)
customer_performing_segements
```

```python
#Leverage Customer Reviews to Drive Trust
sql1_4="""SELECT
    p.ProductName,
    AVG(cr.Rating) AS AvgRating,
    COUNT(cr.ReviewID) AS ReviewCount
    FROM customer_reviews AS cr
    LEFT JOIN products AS p ON cr.ProductID = p.ProductID
    GROUP BY p.ProductName
    ORDER BY AvgRating DESC, ReviewCount DESC"""

my_cursor.execute(sql1_4)
customer_review_trust=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
customer_review_trust=pd.DataFrame(customer_review_trust,columns=columns)
customer_review_trust
```

```python
#Loyalty Programs and Exclusive Offers
sql1_5="""SELECT
    c.CustomerID,
    COUNT(cj.Action) AS PurchaseCount,
    AVG(e.Likes) AS AvgEngagement
    FROM customers AS c
    LEFT JOIN customer_journey AS cj ON c.CustomerID = cj.CustomerID
    LEFT JOIN engagement_data AS e ON c.CustomerID = e.ProductID
    WHERE cj.Action = 'Purchase'
    GROUP BY c.CustomerID
    HAVING PurchaseCount > 3
    ORDER BY AvgEngagement DESC;"""
my_cursor.execute(sql1_4)
loyalty=my_cursor.fetchall()
columns=[col[0] for col in my_cursor.description]
loyalty=pd.DataFrame(loyalty,columns=columns)
loyalty
```
